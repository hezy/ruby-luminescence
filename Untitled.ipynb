```python
import numpy as np
from scipy.special import wofz
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
```

```python
%%latex
Lorenzian:
$$ L(x) = \frac{\gamma}{\sqrt{x^2 + \gamma^2}} $$
```

```python
def lorentz(x, gamma):
    return gamma/np.pi/(np.square(x) + np.square(gamma)) 
```

```python
%%latex
Gaussian:
$$ G(x) = \frac{1}{\sqrt{2 \pi} \sigma} e^{-\frac{x^2}{2 \sigma^2}} $$
```

```python
def gauss(x, sigma):
    return 1/(np.sqrt(2*np.pi)*sigma) * (np.exp(-np.square(x)/2/np.square(sigma)))
```

```python
%% La
Voigt profile is de
```

```python
def voigt(x, sigma, gamma):
    return np.real(wofz((x + 1j*gamma)/(sigma * np.sqrt(2)))) / (sigma * np.sqrt(2*np.pi))
```

```python
x = np.arange (-12.0, 12.0 , 0.01)
yL = lorentz(x,1.0)
yG = gauss(x,1.0)
yV = voigt(x, 1.0, 1.0)
```

```python
plt.close('all')
fig, ax = plt.subplots(figsize=(14, 8))
ax.grid(b=True, which='both', axis='both')
ax.minorticks_on()
ax.set_title("Gaussian, Lorentzian and Voiget", fontsize=16)
ax.set_xlabel("x", fontsize=14)
#ax.set_xlim()
ax.set_ylabel("y", fontsize=14)
#ax.set_ylim()
ax.plot(x,yL,'-r')
ax.plot(x,yG,'-b')
ax.plot(x,yV,'-g')
plt.show()
```
